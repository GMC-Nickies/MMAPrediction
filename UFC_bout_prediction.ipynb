```python
x_data = UFC_data.drop('Winner', axis=1)
x_data = x_data.drop('finish', axis=1)
```

```python
pd.set_option('display.max_columns', None)
print(x_data)
```

```python
labels = UFC_data['Winner']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data,labels,test_size=0.33, random_state=101)
```

```python
print(X_train)
print(X_test)
```

```python
model = LogisticRegression(solver='lbfgs')
X1_train = X_train.drop('R_fighter', axis=1)
X1_train = X1_train.drop('B_fighter', axis=1)
X1_train = X1_train.drop('weight_class', axis=1)
X1_train = X1_train.drop('R_Stance', axis=1)
X1_train = X1_train.drop('B_Stance', axis=1)
rfe = RFE(model, n_features_to_select=24)
fit = rfe.fit(X1_train, y_train)
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
input_func = tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=12,
                                                           num_epochs=3000,shuffle=True)
```

```python
dnn_model = tf.compat.v1.estimator.DNNClassifier(hidden_units=[30,30,30],feature_columns=feat_cols,n_classes=2)
```

```python
dnn_model.train(input_fn=input_func,steps=1000)
```

```python
eval_input_func = tf.compat.v1.estimator.inputs.pandas_input_fn(
      x=X_test,
      y=y_test,
      batch_size=10,
      num_epochs=1,
      shuffle=False)
```

```python
results = dnn_model.evaluate(eval_input_func)
```
